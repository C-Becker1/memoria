```python
import matplotlib.pyplot as plt
from sklearn import preprocessing
import numpy as np
```

```python

def plot_comparacion_diversidad(data):
    # data[0:2] => Misma categoría, diferentes alphas. Con control de diversidad
    # fig, axs = plt.subplots(2,1, sharex=True, sharey=True)
    scaler = preprocessing.MinMaxScaler()

    plt.plot(scaler.fit_transform( np.array( data[0]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[1]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[2]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[3]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[4]).reshape(-1,1) ))
    # axs[0].set_ylabel("Diversidad\n control")

    # data[3:5] => Misma categoría, diferentes alphas. Sin control de diversidad
    # axs[1].plot(scaler.fit_transform( data[3] ))
    # axs[1].plot(scaler.fit_transform( data[4] ))
    # axs[1].plot(scaler.fit_transform( data[5] ))
    # axs[1].set_ylabel("Diversidad\n no control")
    
    plt.legend(["Categoría 1","Categoría 2", "Categoría 3", "Categoría 4", "Categoría 5"])
    plt.xlabel("Generación")
    plt.ylabel("Diversidad")
    plt.show()

def plot_comparacion_diversidad_2(data):
    """
    Grafica la evolución de la diversidad con y sin control para cada valor de alpha.
    Por lo tanto, se obtienen 3 graficos.
    """
    fig, axs = plt.subplots(3,1, sharex=True, sharey=True)
    scaler = preprocessing.MinMaxScaler()

    # data[0:1] => Misma categoría, mismo alpha. Con y sin control
    axs[0].plot(scaler.fit_transform( np.array(data[0]).reshape(-1,1) ))
    axs[0].plot(scaler.fit_transform( np.array(data[1]).reshape(-1,1) ))
    axs[0].plot(scaler.fit_transform( np.array(data[2]).reshape(-1,1) ))
    axs[0].set_ylabel(r"$\alpha=0.3$")

    # data[2:3] => Misma categoría, mismo alpha. Con y sin control
    axs[1].plot(scaler.fit_transform( np.array(data[3]).reshape(-1,1) )) 
    axs[1].plot(scaler.fit_transform( np.array(data[4]).reshape(-1,1) ))
    axs[1].plot(scaler.fit_transform( np.array(data[5]).reshape(-1,1) )) 
    axs[1].set_ylabel("Diversidad \n"+r"$\alpha=0.6$")

    # data[4:5] => Misma categoría, mismo alpha. Con y sin control
    axs[2].plot(scaler.fit_transform( np.array(data[6]).reshape(-1,1) ))
    axs[2].plot(scaler.fit_transform( np.array(data[7]).reshape(-1,1) ))
    axs[2].plot(scaler.fit_transform( np.array(data[8]).reshape(-1,1) ))
    axs[2].set_ylabel(r"$\alpha=0.9$")

    
    fig.legend(["control id", "control hamming", "no control"])
    axs[2].set_xlabel("Generación")
```

```python
def plot_avg_fitness(data):
    """
    Grafica la evolución de la función de aptitud con y sin control para cada valor de alpha.
    Por lo tanto, se obtienen 3 graficos.
    """
    fig, axs = plt.subplots(3,1, sharex=True, sharey=True)
    scaler = preprocessing.MinMaxScaler()

    # data[0:1] => Misma categoría, mismo alpha. Con y sin control
    axs[0].plot(scaler.fit_transform( np.array(data[0]).reshape(-1,1) ))
    axs[0].plot(scaler.fit_transform( np.array(data[1]).reshape(-1,1) ))
    axs[0].plot(scaler.fit_transform( np.array(data[2]).reshape(-1,1) ))
    axs[0].set_ylabel(r"$\alpha=0.3$")

    # data[2:3] => Misma categoría, mismo alpha. Con y sin control
    axs[1].plot(scaler.fit_transform( np.array(data[3]).reshape(-1,1) )) 
    axs[1].plot(scaler.fit_transform( np.array(data[4]).reshape(-1,1) ))
    axs[1].plot(scaler.fit_transform( np.array(data[5]).reshape(-1,1) )) 
    axs[1].set_ylabel("Aptitud \n"+r"$\alpha=0.6$")

    # data[4:5] => Misma categoría, mismo alpha. Con y sin control
    axs[2].plot(scaler.fit_transform( np.array(data[6]).reshape(-1,1) ))
    axs[2].plot(scaler.fit_transform( np.array(data[7]).reshape(-1,1) ))
    axs[2].plot(scaler.fit_transform( np.array(data[8]).reshape(-1,1) ))
    axs[2].set_ylabel(r"$\alpha=0.9$")
    
    axs[2].set_xlabel("Generación")
    
    fig.legend(["control id", "control hamming", "no control"])

def plot_avg_fitness_2(data):
    # data[0:2] => Misma categoría, diferentes alphas. Con control de diversidad
    # fig, axs = plt.subplots(2,1, sharex=True, sharey=True)
    scaler = preprocessing.MinMaxScaler()

    plt.plot(scaler.fit_transform( np.array( data[0]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[1]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[2]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[3]).reshape(-1,1) ))
    plt.plot(scaler.fit_transform( np.array( data[4]).reshape(-1,1) ))
    # axs[0].set_ylabel("Diversidad\n control")

    # data[3:5] => Misma categoría, diferentes alphas. Sin control de diversidad
    # axs[1].plot(scaler.fit_transform( data[3] ))
    # axs[1].plot(scaler.fit_transform( data[4] ))
    # axs[1].plot(scaler.fit_transform( data[5] ))
    # axs[1].set_ylabel("Diversidad\n no control")
    
    plt.legend(["Categoría 1","Categoría 2", "Categoría 3", "Categoría 4", "Categoría 5"])
    plt.xlabel("Generación")
    plt.ylabel("Aptitud")
    plt.show()
    
```
