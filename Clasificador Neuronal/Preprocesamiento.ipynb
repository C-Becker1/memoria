```python
# cargar todas las prendas en un dict por id
"""
{
    id_cloth_1: {
        atributos: (0, 1, 0, 0, ...),
        opiniones: [
            { 
                clima: (1, 0, 0, 1, 0),
                nota: 2,
                id_outfit: 781
            },
            { 
                clima: (1, 1, 0, 1, 0),
                nota: 3,
                id_outfit: 49
            },
            ...
        ]
    },

    id_cloth_2: {
        atributos: (0, 0, 0, 1, ...),
        opiniones: [
            {
                clima: (1, 1, 1, 0, 0),
                nota: 3,
                id_outfit: 325
            },
            ...
        ]
    }
}
"""
```

```python
def integer_encode(attrs):
    i = 0

    patron = 0
    sleeve = 0
    skirt = 0
    neckline = 0
    material = 0
    fit = 0

    while i < len(attrs):
        if int(attrs[i]) == 1:

            if 0 <= i <= 6: # pattern design
                patron = i + 1
            if 7 <= i <= 9: # sleeve length
                sleeve = i - 6
            if 10 <= i <= 12: # skirt length
                skirt = i - 9
            if 13 <= i <= 16: # neckline
                neckline = i - 12
            if 17 <= i <= 22: # material
                material = i - 16
            if 23 <= i <= 25: # fit 
                fit = i - 22
        i += 1
    
    return patron, sleeve, skirt, neckline, material, fit

def unpack_cloth(cloth_string):
    cloth_string = cloth_string.replace('"', '')
    cloth_list = cloth_string.split(",")
    # print(cloth_list)
    id_cloth, name, description, img_route, cloth_category, cloth_subcategory = cloth_list[0:6]
    name = name.replace("-img", "")

    terceraCapa = ["anorak", "bomber", "jacket", "parka", "peacoat", "poncho"]
    segundaCapa = ["blazer", "cardigan", "hoodie", "jersey", "sweater", "turtleneck", "flannel"]
    primeraCapa = ["blouse", "button-down", "halter", "henley", "tank", "tee", "top", "turtleneck", "flannel"]

    attributes = cloth_list[6:]
    patron, sleeve, skirt, neckline, material, fit = integer_encode(attributes)
    cloth = {
        "name": name,
        # "description": description,
        "img_route": img_route, 
        "cloth_category": cloth_category,
        "cloth_subcategory": cloth_subcategory,
        "patron": patron,
        "sleeve": sleeve,
        "skirt": skirt,
        "neckline": neckline,
        "material": material,
        "fit": fit,
        "opinions": []
    }
    return id_cloth, cloth

def read_cloths_csv(filename):
    file = open(filename)
    clothes = {}

    for line in file:
        line = line.strip()
        id_cloth, cloth = unpack_cloth(line)
        if id_cloth not in clothes:
            clothes[id_cloth] = cloth
    file.close()
    return clothes
```

#### Wisdom of the Crowd
Bajo el concepto "wisdom of the crowd" o sabiduría de los grupos, que indica que una decisión es más acertada cuando se considera la opinión del grupo por sobre una opinión individual, es que se decide utilizar la moda como métrica para definir las categorías de clima de una prenda

#### Promedio
Para obtener la nota de una prenda, se calcula el promedio de ésta, se redondea y se convierte en un entero

```python
    def list_divition(list, denominator):
        retorno = []
        for elem in list:
            retorno.append( elem / denominator)
        return retorno

    def simplificar(opinions):
        # print(opinions)
        climas = [0, 0, 0, 0, 0]
        suma = 0
        for o in opinions:
            weather = o["climas"].split(",")
            for w in weather:
                climas[int(w) - 1] += 1
            
            suma += int(o["nota"])
        
        climas = list_divition(climas, len(opinions))

        i = 0
        while i < len(climas):
            if climas[i] >= 0.5:
                climas[i] = 1
            else:
                climas[i] = 0
            i += 1

        promedio = suma / len(opinions)
        nota = int(round(promedio))
        print(climas)
        return climas, nota
```

```python
def getCategories(category, subcategory):
    subcategory = subcategory.lower()

    if category == "UPPER BODY":
        if subcategory in primeraCapa:
            category_number = 1
            subcategory_number = primeraCapa.index(subcategory) + 1
        elif subcategory in segundaCapa:
            category_number = 2
            subcategory_number = segundaCapa.index(subcategory) + 1
        elif subcategory in terceraCapa:
            category_number = 3
            subcategory_number = terceraCapa.index(subcategory) + 1
    
    elif category == "HAT":
        category_number = 4
        subcategory_number = hats_subcategories.index(subcategory) + 1
    
    elif category == "SHOES":
        category_number = 5
        subcategory_number = shoes_subcategories.index(subcategory) + 1 

    elif category == "LOWER BODY":
        category_number = 6
        subcategory_number = lower_body_subcategories.index(subcategory) + 1
    
    return category_number, subcategory_number
```

```python
def dict_to_string(key, value, separation = ","):
    values = [str(key)]
    for k in value:
        values.append(str(value[k]))
    return separation.join(values)


def write_csv_file(filename, column_names, data):
    file = open(filename, "w")
    file.write(",".join(column_names) + "\n")
    
    for k in data:
        file.write(f"{dict_to_string(k, data[k])}\n")
    
    file.close()

column_names = []
for k in clothes:
    # del clothes[k]["name"]
    # del clothes[k]["img_route"]
    for attr in clothes[k]:
        column_names.append(attr)
    break

write_csv_file("clothes.csv", column_names, clothes)
        
```

```python
def read_cloth_outfit_relation(filename):
    file = open(filename)
    outfits = {}
    for line in file:
        line = line.strip()
        line = line.replace('"', '')
        id, id_outfit, id_cloth = line.split(",")
        id_outfit = int(id_outfit)
        id_cloth = int(id_cloth)

        if id_outfit not in outfits:
            outfits[id_outfit] = []
        
        if id_cloth not in clothes:
            outfits[id_outfit].append( {"cloth_category": -1} )
        else:
            outfits[id_outfit].append( clothes[id_cloth] )
    file.close()
    return outfits
```

```python
def decodeWeather(string):
    string = string.split(",")
    weather = ["0", "0", "0", "0", "0"]
    for w in string:
        weather[int(w)-1] = "1"
    return weather


def read_outfit_opinion(nomarch):
    evaluated_outfits = []
    file = open(nomarch)
    outfit_file = open("outfit_target.csv", "w")
    categories = ["isCold", "isRainy", "isWet", "isTempered", "isHot", "nota"]
    outfit_file.write(",".join(categories) + "\n")

    data = []

    for line in file:
        line = line.strip()
        _, id_outfit, _, clima, nota = unpack_cloth_opinion(line)
        if int(id_outfit) > 50 and int(id_outfit) not in evaluated_outfits:
            evaluated_outfits.append(int(id_outfit))
            
            target = decodeWeather(clima)
            target.append(nota)
            target.insert(0, int(id_outfit))
            data.append(target)
        # outfit_file.write(",".join(target) + "\n")
    

    data.sort()

    for d in data:
        
        d[0] = str(d[0])
        outfit_file.write(",".join(d) + "\n")

    
    file.close()
    outfit_file.close()
    
    return evaluated_outfits

evaluated_outfit = read_outfit_opinion("../../data_muestreada/outfit_opinion.csv")
```

```python
def cloth_in_outfit(outfit, cloth_category):
    for cloth in outfit:
        if cloth["cloth_category"] == cloth_category:
            return cloth
    return False

def dict_to_string_2(dict, separator=","):
    array = []
    for k in dict:
        array.append(str(dict[k]))
    return separator.join(array)

column_names = []
for k in clothes:
    for attr in clothes[k]:
        column_names.append(attr)
    break

outfit_strings = []

empty_cloth = {}
for attr in column_names:
    empty_cloth[attr] = 0

for id_outfit in outfits:
    if id_outfit in evaluated_outfit:
        outfit_string = [str(id_outfit)]
        for i in [1,2,3,4,5,6]:
            cloth = cloth_in_outfit(outfits[id_outfit], i) 
            if cloth:
                outfit_string.append( dict_to_string_2(cloth) )
            else:
                outfit_string.append( dict_to_string_2(empty_cloth) )
        outfit_string = ",".join(outfit_string)
        outfit_strings.append(outfit_string)

            
```
